```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from glob import glob
import os
import matplotlib.gridspec as gridspec
import matplotlib.ticker as ticker
sns.set_style('whitegrid')
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

#from generator import DataGenerator
import tensorflow.keras
```

```python
# دا بيقرأ الملف ... علشان فيه بيانات الصور اللي موجوده عندي 
all_xray_df = pd.read_csv('/content/dataset/data/Data_Entry_2017_v2020.csv')
# دا بيخزن ليست من المسارات للصور 
#this function returns a dictionary ,prepair it to be passed to a map function 
all_image_paths = {os.path.basename(x): x for x in 
                   glob(os.path.join( '/content/images/*.png'))}


#سطر طباعه عادي 
#print('Scans found:', len(all_image_paths), ', Total Headers', all_xray_df.shape[0])
#دا بيعمل عمود جديد ف الملف و بيضيف فيه كل مسار مع اسم الصوره علي حسب ال LABEL 
#the map function here is applied to a a pd series henc every column in a dataframe is a series . 
#the map here takes a dictionary instead of a function , and the pd series as the input .
all_xray_df['path'] = all_xray_df['Image Index'].map(all_image_paths.get)

all_xray_df['path'] = all_xray_df['path'].astype(str)

#all_xray_df['Patient Age'] = all_xray_df['Patient Age'].map(lambda x: int(x[:-1]))
all_xray_df.sample(3)
```

```python
#بنعرض الداتا علشان نشوف مالها لقيناها ان هي biased 
label_counts = all_xray_df['Finding Labels'].value_counts()[:15]
fig, ax1 = plt.subplots(1,1,figsize = (12, 8))
ax1.bar(np.arange(len(label_counts))+0.5, label_counts)
ax1.set_xticks(np.arange(len(label_counts))+0.5)
_ = ax1.set_xticklabels(label_counts.index, rotation = 90)
```

```python
#here we have some diseases have label of "finding label" , replace it with empty space  
all_xray_df['Finding Labels'] = all_xray_df['Finding Labels'].map(lambda x: x.replace('No Finding', ''))
 

from itertools import chain
# to display the different labels 
all_labels = np.unique(list(chain(*all_xray_df['Finding Labels'].map(lambda x: x.split('|')).tolist())))
all_labels = [x for x in all_labels if len(x)>0]
print('All Labels ({}): {}'.format(len(all_labels), all_labels))

for c_label in all_labels:
    print(len(c_label))
    if len(c_label)>1:
        all_xray_df[c_label] = all_xray_df['Finding Labels'].map(lambda finding: 1.0 if c_label in finding else 0)
all_xray_df.sample(3)
```

```python
 # nothing to say , easy 
MIN_CASES = 1000
#هات كل مرض عندك و عدد الحالات
all_labels = [c_label for c_label in all_labels if all_xray_df[c_label].sum()>MIN_CASES]
print('Clean Labels ({})'.format(len(all_labels)), 
      [(c_label,int(all_xray_df[c_label].sum())) for c_label in all_labels])
```

```python
# هنا عرضنا كل عدد حالاات كل مرض 
# since the dataset is very unbiased, we can resample it to be a more reasonable collection

sample_weights = all_xray_df['Finding Labels'].map(lambda x: len(x.split('|')) if len(x)>0 else 0).values + 4e-2
print(sample_weights)
#scaling the weights
sample_weights /= sample_weights.sum()
print(sample_weights)
all_xray_df = all_xray_df.sample(40000, weights=sample_weights)
print(all_xray_df)

label_counts = all_xray_df['Finding Labels'].value_counts()[:15]
fig, ax1 = plt.subplots(1,1,figsize = (12, 8))
ax1.bar(np.arange(len(label_counts))+0.5, label_counts)
ax1.set_xticks(np.arange(len(label_counts))+0.5)
_ = ax1.set_xticklabels(label_counts.index, rotation = 90)
```

```python
label_counts = 100*np.mean(all_xray_df[all_labels].values,0)
print(label_counts)
fig, ax1 = plt.subplots(1,1,figsize = (12, 8))
ax1.bar(np.arange(len(label_counts))+0.5, label_counts)
ax1.set_xticks(np.arange(len(label_counts))+0.5)
ax1.set_xticklabels(all_labels, rotation = 90)
ax1.set_title('Adjusted Frequency of Diseases in Patient Group')
_ = ax1.set_ylabel('Frequency (%)')
```

```python
# دا هيضيف عمود جديد هيتحط فيه فيكتور مكون من اصفار و وحايد الواحد بيعبر عن ان الصوره دي فيها المرض كذا و الصفر معناها مفيهوش المرض دا 
all_xray_df['disease_vec'] = all_xray_df.apply(lambda x: [x[all_labels].values], 1).map(lambda x: x[0])
all_xray_df.sample(3)
```

```python
from sklearn.model_selection import train_test_split
train_df, valid_df = train_test_split(all_xray_df, 
                                   test_size = 0.20, 
                                   random_state = 2018,
                                   stratify = all_xray_df['Finding Labels'].map(lambda x: x[:4])
                                     )
print('train', train_df.shape[0], 'validation', valid_df.shape[0])
```

```python
valid_df['newLabel'] = valid_df.apply(lambda x: x['Finding Labels'].split('|'), axis=1)
train_df['newLabel'] = train_df.apply(lambda x: x['Finding Labels'].split('|'), axis=1)
train_df.sample(3)
```

```python
from keras.preprocessing.image import ImageDataGenerator
IMG_SIZE = (128, 128)
core_idg = ImageDataGenerator()
core_idg_2 = ImageDataGenerator(samplewise_center=True, 
                              samplewise_std_normalization=True, 
                              horizontal_flip = True, 
                              vertical_flip = False, 
                              height_shift_range= 0.05, 
                              width_shift_range=0.1, 
                             rotation_range=5, 
                              shear_range = 0.1,
                              zoom_range=0.15)
```

```python
train_gen = core_idg_2.flow_from_dataframe(dataframe=train_df, 
                             directory=None,
                             x_col = 'path',
                            y_col = 'newLabel', 
                            classes = all_labels,
                            target_size = IMG_SIZE,
                             color_mode = 'rgb',
                            batch_size = 16)

valid_gen = core_idg_2.flow_from_dataframe(dataframe=valid_df, 
                             directory=None,
                             x_col = 'path',
                            y_col = 'newLabel', 
                            classes = all_labels,
                            target_size = IMG_SIZE,
                             color_mode = 'rgb',
                            batch_size = 32) 

test_X, test_Y = next(core_idg.flow_from_dataframe(dataframe=valid_df, 
                             directory=None,
                             x_col = 'path',
                            y_col = 'newLabel', 
                            classes = all_labels,
                            target_size = IMG_SIZE,
                             color_mode = 'rgb',
                            batch_size = 8000))
```

```python
t_x, t_y = next(train_gen)
fig, m_axs = plt.subplots(4, 4, figsize = (16, 16))
for (c_x, c_y, c_ax) in zip(t_x, t_y, m_axs.flatten()):
    c_ax.imshow(c_x[:,:,0])
    c_ax.set_title(', '.join([n_class for n_class, n_score in zip(all_labels, c_y) 
                             if n_score>0.5]))
    c_ax.axis('off')
```

```python
from keras.applications.densenet import preprocess_input
from tensorflow.keras.layers import Input
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense
#from keras.optimizers import Adam
```

```python
# dense net model
img_in = Input(t_x.shape[1:])              
model = DenseNet121(include_top= False , 
                weights="imagenet",     
                input_tensor= img_in, 
                input_shape= t_x.shape[1:],
                pooling ='avg') 

x = model.output  
#softmax
pre = Dense(len(all_labels), activation="sigmoid", name="pre")(x)   
model = Model(inputs=img_in, outputs=pre)
```

```python
from tensorflow import keras
optimizer = Adam(lr=1e-1)
model.compile(optimizer=optimizer, loss="binary_crossentropy", metrics=[keras.metrics.binary_accuracy])
```

```python

from tensorflow.keras.callbacks import EarlyStopping,ReduceLROnPlateau
early = EarlyStopping(monitor="loss", mode="min",min_delta = 0,
                          patience = 3,
                          verbose = 1,
                          restore_best_weights = True)
learning_rate_reduction = ReduceLROnPlateau(monitor='loss', patience = 2, verbose=1,factor=0.3, min_lr=0.000001)

callbacks_list = [ early, learning_rate_reduction]
```

```python
history = model.fit_generator(train_gen, 
                                 
                                  validation_steps=1,
                                  validation_data = valid_gen, 
                                  epochs = 1000 , callbacks = callbacks_list)
```

```python
y_pred = model.predict(test_X) 
```
